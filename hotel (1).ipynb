```python
class client:
    def __init__(self,name,phone_number):
        self.name=name
        self.phone_number=phone_number

    def __str__(self):
        return f"{self.name} {self.phone_number}"

class room:
    def __init__(self,room_number,room_type):
        self.room_number=room_number
        self.room_type=room_type
        self.available=True

    def __str__(self):
        if self.available==True:
          return f"the {self.room_type} room {self.room_number} is available"
        elif self.available==False:
          return f"the {self.room_type} room {self.room_number} is not available"
        else:
          raise ValueError("this room does not exist")

class hotel:
    def __init__(self):
        self.rooms=[]
        self.clients=[]
        for i in range(101,110):
            self.rooms.append(room(i,"single"))
        for i in range(201,210):
            self.rooms.append(room(i,"double"))
        for i in range(301,310):
            self.rooms.append(room(i,"family"))
        for i in range(401,410):
            self.rooms.append(room(i,"suite"))

    def reserve_room(self,room_number,client):
        for room in self.rooms:
            if room.room_number==room_number:
                if room.available==True:
                    room.available=False
                    self.clients.append(client)
                    print("the room has been successfully reserved")
                    break
                elif room.available==False:
                    raise ValueError("the room is not available")
                    break
                else:
                    raise ValueError("this room does not exist")
                    break

    def show_rooms(self):
        for room in self.rooms:
             if room.available==True:
                print(f"the {room.room_type} room {room.room_number} is available")

    def cancel_booking(self):
      room_number=int(input("enter the room number: "))
      for room in self.rooms:
        if room.room_number==room_number:
          if room.available==True:
              raise ValueError("this room is not yet booked")
          elif room.available==False:
              room.available=True
              print("booking sucessfully cancelled")
          else:
              raise ValueError("this room does not exist")
              break

    def menu(self):
      while True:
        print(""" options:
        1. show rooms
        2. reserve room
        3. cancel booking
        4. exit""")
        option=int(input("enter your option: "))
        if option==1:
          self.show_rooms()
        elif option==2:
          name=input("enter your name: ")
          phone_number=int(input("enter your phone number: "))
          current_client=client(name,phone_number)
          room_number=int(input("enter the room number: "))
          self.reserve_room(room_number,current_client)
        elif option==3:
          self.cancel_booking()
        elif option==4:
          break
        else:
          print("invalid option")

hotel=hotel()
hotel.menu()
```
